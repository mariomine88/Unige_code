Compute the FFT
---------------
Using the ${\tt numpy}$ library, compute:

<ul>
    <li>the FFT</li>
    <li>its matching vector of frequencies (np.fft.fftfreq)</li>
    <li>the inverse FFT (back to the signal)</li>
</ul>

```python
n = len(time) ## number of samples

## array of fft(y)
fft_y = np.fft.fft(y)

## array of frequencies values
freq = np.fft.fftfreq(n, 1/sampling_r)

## estimate y through inverse FFT
y_est = np.fft.ifft(fft_y)
```

## Verify Parseval

Now we want to verify that the Parseval euqualiti holds. We first compute the energy of the signal y, then we compute the energy of the fft(y). 

We normalize once w.r.t. to the number of samples following <a href="https://en.wikipedia.org/wiki/Discrete_Fourier_transform#The_Plancherel_theorem_and_Parseval's_theorem" target="_blank">this notation</a>. 

```python
def squared_norm(func):
    return np.vdot(func,func)
```

```python
## parseval - norm of the function

## to evaluate the inner product we make use of the squared_norm function
norm_y = squared_norm(y)

print(norm_y)
```

```python
## parseval - sum of the squared coefficients of the series

## to evaluate the inner product we make use of the squared_norm function

norm_fft_y = 1/n*squared_norm(fft_y)

print(np.abs(norm_fft_y))
```

```python
norm_y_est = squared_norm(y_est)
print(np.abs(norm_y_est))
```

```python
sampling_rates = [100, 50, 20, 10]  # Decreasing sampling rates

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for idx, sampling_r_test in enumerate(sampling_rates):
    # Regenerate the signal with the new sampling rate
    time1_test = np.arange(-10, 0, 1/sampling_r_test)
    time2_test = np.arange(0.0, 10, 1/sampling_r_test)
    time3_test = np.arange(10, 20, 1/sampling_r_test)
    
    y1_test = time1_test * 0
    y2_test = time2_test * (time2_test - a) * np.sin(30 * time2_test * np.pi / a)
    y3_test = time3_test * 0
    
    time_test = np.concatenate([time1_test, time2_test, time3_test])
    y_test = np.concatenate([y1_test, y2_test, y3_test])
    
    # Compute FFT
    fft_y_test = np.fft.fft(y_test)
    N_test = len(y_test)
    freq_test = np.fft.fftshift(np.fft.fftfreq(N_test, 1/sampling_r_test))
    fft_y_test_shifted = np.fft.fftshift(fft_y_test)
    
    # Plot
    axes[idx].plot(freq_test, np.abs(fft_y_test_shifted))
    axes[idx].set_xlim([-50, 50])
    axes[idx].set_xlabel('Frequency (Hz)')
    axes[idx].set_ylabel('|FFT(y)|')
    axes[idx].set_title(f'Sampling Rate = {sampling_r_test} Hz (Nyquist = {sampling_r_test/2} Hz)')
    axes[idx].grid(True)

plt.tight_layout()
plt.show()
```
