Nonostante le differenze tra le due tabelle, la funzione concat concatena le due tabelle, mettendo a Null i valori delle colonne presenti solo in una delle due ed in corrispondenza delle righe provenienti dal dataset in cui la colonna non è presente. Proviamo a verificare la presenza di Null

```python
# ... completare qui ...

integrate_titles.isnull().sum()
```

Ad esempio la colonna **listed_in**, che è presente solo in netflix_titles, ha 1070 valori Null, che corrisponde al numero di righe della tabella disney_titles.

Decidiamo di eliminare eventuali colonne non comuni. Possiamo farlo **prima** dell'integrazione, oppure direttamente sulla tabella concatenata.

**Tip**: confrontate i due schemi per capire quali siano le colonne non comuni. Attenzione!!! Potrebbero anche esserci colonne che sembrano non comuni, ma in realtà contengono la stessa tipologia di informazione e hanno solo nomi diversi!!

```python
# ... completare qui ...

# Nota:
# listed_in è su netflix ma non disney
# description è su netflix ma non disney
# cast è su netflix ma non disney
# type e show_type sono la stessa cosa
# date_added in netflix e from in Disney sono la stessa cosa

integrate_titles=integrate_titles.drop(columns=["listed_in"], axis=1)
integrate_titles=integrate_titles.drop(columns=["description"], axis=1)
integrate_titles=integrate_titles.drop(columns=["cast"], axis=1)

integrate_titles.info()
```

Andiamo ora alla ricerca di eventuali colonne che contengono la stessa informazione. Guardate in particolare (type, show_type),  (contry, production_countries), (date_added, from)

```python
# ...completare qui ...

# dato che type e show_type sono due colonne che contengono la stessa informazione e hanno solo nome diverso,
# dobbiamo integrate le due colonne in una singola, buttando via l'altra
# Copiare la porzione non nulla di show_type dentro a type

# Anche date_added e from sono la stessa cosa, vale la cosa di prima
import numpy as np

integrate_titles.iloc[integrate_titles['show_type'].isna().index,integrate_titles.columns.get_loc('show_type')] =  integrate_titles.iloc[integrate_titles['show_type'].isna().index,integrate_titles.columns.get_loc('type')]

integrate_titles.iloc[np.where(integrate_titles['date_added'].isna())[0],integrate_titles.columns.get_loc('date_added')] =  integrate_titles.iloc[np.where(integrate_titles['date_added'].isna())[0],integrate_titles.columns.get_loc('from')]

integrate_titles=integrate_titles.drop(columns=["type"], axis=1)
integrate_titles=integrate_titles.drop(columns=["from"], axis=1)

# contry e production_country sono ridondanti, possiamo buttarne via una
```

```python
print(integrate_titles['date_added'].isna().sum())
print(integrate_titles['show_type'].isna().sum())
#
```

Infine, ci sono colonne che già ora possiamo giudicare inutili?

```python
# ...completare qui ...

integrate_titles=integrate_titles.drop(columns=["Unnamed: 0"], axis=1)

integrate_titles.info()
```

### Carichiamo altre due tabelle, che forniscono fonti di informazioni aggiuntive

```python
netflix_credits = pd.read_csv("netflix_credits.csv")
disney_credits = pd.read_csv("disney_credits.csv")
```

come sempre, stampiamo le prime righe di ciascun dataFrame

```python
netflix_credits.head()
```

Possiamo osservare che le tabelle credits e titles, hanno informazioni complementari.  In particolare, credits riporta delle informazinoi aggiuntive riguardanti attori e carattere interpretato.

Come prima, procediamo con l'integrazione delle due tabelle, che questa volta è più facile (lo schema è proprio lo stesso!)

```python
integrate_credits = pd.concat([netflix_credits,disney_credits])
```

Data la complementarietà delle informazioni che troviamo nelle cartelle titles e credits, possiamo decidere di fare un **join** tra le due tabelle, per includere tutte le informazioni in un unico dataset.


```python
# ...completare qui usando la funzione merge per fare il join
final_table=pd.merge(integrate_titles,
         integrate_credits,
         how = "inner",
         left_on = "id",
         right_on = "id");
```

verifichiamo cosa sia successo guardando le dimensioni di ciascun dataframe

```python
print(integrate_titles.shape)
print(integrate_credits.shape)
print(final_table.shape)
```

concentriamoci sulla colonna date_added, intanto visualizzandola:

```python
print(final_table['date_added'])
```

vediamo che giorni mesi e anni sono presenti. Potrebbe essere utile, però, separare queste informazioni, incorporandole in tre colonne diverse.

Facciamolo insieme, con due diversi passi:

1. dobbiamo trasformare questa colonna in formato 'datetime' utilizzando la funzione di pandas 'to_datetime'

```python
final_table['date_added'] = pd.to_datetime(final_table['date_added'])
```

```python
print(final_table['date_added'])
```

poi, possiamo separare le varie entrate in tre nuove colonne, che si chiameranno day, month e year.

```python
final_table['year'] = final_table['date_added'].dt.year
final_table['month'] = final_table['date_added'].dt.month
final_table['day'] = final_table['date_added'].dt.day
```

ora visualizziamo il risultato

```python
final_table.head()
```

ora creiamo una copia della tabella, e modifichiamo la durata dei film da minuti a ore

Attenzione, ci sono dei campioni in cui la durata è in season, ma il tipo è MOVIE.
Questi possono essere eliminati

```python
final_table=pd.merge(integrate_titles,
         integrate_credits,
         how = "inner",
         left_on = "id",
         right_on = "id");
movies_indexes = np.where(final_table['show_type']=='MOVIE')[0]
#troviamo questi campioni ed eliminiamoli
print(np.sum(final_table.iloc[movies_indexes,final_table.columns.get_loc('duration')].str.contains('Season')))
final_table = final_table.drop(final_table[final_table['duration'].str.contains('Season', na=False)].index)
movies_indexes = np.where(final_table['show_type']=='MOVIE')[0]
print(np.sum(final_table.iloc[movies_indexes,final_table.columns.get_loc('duration')].str.contains('Season')))
```

a questo punto, possiamo modificare la durata in ore, per prima cosa eliminiamo la parola minuti

```python
final_table['duration']=final_table['duration'].replace(' min','',regex=True)
```

quindi dividere per 60 (dopo conversione float)

```python
print(final_table.iloc[movies_indexes,final_table.columns.get_loc('duration')])
```

e finalmente, aggiungere un suffisso per indicare le ore, per esempio 'hours', tornando a stringa

```python
final_table.iloc[movies_indexes,final_table.columns.get_loc('duration')] = final_table.iloc[movies_indexes,final_table.columns.get_loc('duration')].astype(float)/60
final_table.iloc[movies_indexes,final_table.columns.get_loc('duration')] = final_table.iloc[movies_indexes,final_table.columns.get_loc('duration')].astype(str).str[:5]+ ' hours'
```

stampiamo il risultato finale

```python
print(final_table['duration'])
```
